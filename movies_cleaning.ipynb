```python
import pandas as pd 
import numpy as np
import ast

df = pd.read_csv(r'F:\Projects\Internship\Technocolab\movie_ratings\movies_metadata.csv')
movies = pd.DataFrame()
cols_to_drop = ['adult','imdb_id','revenue', 'overview', 'spoken_languages', 'status', 'tagline', 'original_title', 'video', 'vote_count']
df = df.drop(cols_to_drop, axis = 1)
```

```python
movies['movie_id'] = df['id']
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['belongs_to_collection'] = df['belongs_to_collection'].fillna('[]').apply(ast.literal_eval)
df['production_companies'] = df['production_companies'].fillna('[]').apply(ast.literal_eval)
df['production_countries'] = df['production_countries'].fillna('[]').apply(ast.literal_eval)
```

```python
for i in range(df.shape[0]) : 
    try : 
        movies.loc[i,'collection'] = df.loc[i, 'belongs_to_collection']['name']
    except : 
        movies.loc[i,'collection'] = np.nan
```

```python
for i in range(df.shape[0]):
    temp = ast.literal_eval(df['genres'][i])
    try : 
        movies.loc[i, 'genre1'] = temp[0]['name']
    except :
         movies.loc[i, 'genre1'] = np.nan
    try : 
        movies.loc[i, 'genre2'] = temp[1]['name']
    except :
         movies.loc[i, 'genre2'] = np.nan
```

```python
for i in range(df.shape[0]):
    temp = df.loc[i,'production_countries']
    
    try : 
        movies.loc[i, 'production_country_1'] = temp[0]['name']
    except :
         movies.loc[i, 'production_country_1'] = np.nan
    try : 
        movies.loc[i, 'production_country_2'] = temp[1]['name']
    except :
         movies.loc[i, 'production_country_2'] = np.nan
```

```python
for i in range(df.shape[0]):
    temp = df.loc[i,'production_companies']

    try : 
        movies.loc[i, 'production_company_1'] = temp[0]['name']
    except :
         movies.loc[i, 'production_company_1'] = np.nan
    try : 
        movies.loc[i, 'production_company_2'] = temp[1]['name']
    except :
         movies.loc[i, 'production_company_2'] = np.nan
```

```python
movies.head()
```

```python
df['budget'] = df['budget'].replace('0',np.nan)

remaining_var = ['budget', 'title', 'original_language', 'popularity', 'homepage', 'poster_path', 'runtime', 'vote_average']

for i in remaining_var : 
    movies[i] = df[i]
```
